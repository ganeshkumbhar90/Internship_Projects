# Question - 2:

```python
# Here we have extracted the "urls":

# Used 3 here, because we have to extract the data till page 3.

start = 0
end = 3 

url = []
for i in range(start,end):
    try:
        ur = driver.find_elements_by_xpath('//h2[@class="a-size-mini a-spacing-none a-color-base s-line-clamp-2"]/a')
        for i in ur:
            url.append(i.get_attribute('href'))       
    except NoSuchElementException:
        url.append("-")
    
    time.sleep(2)
    nxt_button = driver.find_elements_by_xpath("//div[@class='a-section a-spacing-large a-spacing-top-large a-text-center s-pagination-container']/span/a[3]")   
    try:
        driver.get(nxt_button[0].get_attribute('href')) 
    except:
        driver.get(nxt_button[1].get_attribute('href'))
len(url)
```

```python
# Creation the empty lists:
brand=[]     
price=[]
product=[]
exchange=[]
delivery=[]
availability=[]
# Used try and except:
for i in url:    
    try:                           
        driver.get(i)
        time.sleep(2)             
        br=driver.find_element_by_xpath('//h1[@class="a-size-large a-spacing-none"]')
        brand.append(br.text)                                  
    except NoSuchElementException:
        brand.append("-")
        
    try:
        pri=driver.find_element_by_xpath('//span[@class="a-price-whole"]')
        price.append(pri.text)
    except NoSuchElementException:
        price.append("-")
        
    try:
        pro=driver.find_element_by_xpath('//span[@class="a-size-large product-title-word-break"]')
        product.append(pro.text)
    except NoSuchElementException:
        product.append("-")
    
    try:
        exc=driver.find_element_by_xpath('//*[@id="RETURNS_POLICY"]/span/div[2]')
        exchange.append(exc.text)
    except NoSuchElementException:
        exchange.append("-")
    
    try:
        de=driver.find_element_by_xpath('//*[@id="ddmDeliveryMessage"]/b')
        delivery.append(de.text)
    except NoSuchElementException:
        delivery.append("-")
    
    try:
        ava=driver.find_element_by_xpath('//span[@class="a-size-medium a-color-success"]')
        availability.append(ava.text)
    except NoSuchElementException:
        availability.append("-")
    
        
print(len(brand))              
print(len(price)) 
print(len(product))
print(len(exchange))
print(len(delivery))
print(len(availability))
```

```python
# Creating the DataFrame - 
df = pd.DataFrame({'Brand Name':brand,'Name of the Product':product,'Price':price,'Return/Exchange':exchange,'Expected Delivery':delivery,'Availability':availability,'Product URL':url})
df
```

```python
df. to_csv("Products.csv")
```

```python
brand=[]                                
price=[]
product=[]
colour=[]
ram=[]
rom=[]
primary=[]
secondary=[]
display=[]
battery=[]

for i in url:

    try:                                
        driver.get(i)
        time.sleep(2)
        bran=driver.find_element_by_xpath('//span[@class="B_NuCI"]')
        brand.append(bran.text.split()[0])
        but=driver.find_element_by_xpath('//button[@class="_2KpZ6l _1FH0tX"]')
        but.click()
    except NoSuchElementException:
        brand.append("-")
        
    try:
        pric=driver.find_element_by_xpath('//div[@class="_30jeq3 _16Jk6d"]')
        price.append(pric.text)
    except NoSuchElementException:
        price.append("-")
        
    try:
        prod=driver.find_element_by_xpath('//span[@class="B_NuCI"]')
        product.append(prod.text)
    except NoSuchElementException:
        product.append("-")
        
    try:
        colo=driver.find_element_by_xpath('//span[@class="B_NuCI"]')
        colour.append(colo.text.split()[2:4])
    except NoSuchElementException:
        colour.append("-")
    
    try:
        ra=driver.find_element_by_xpath('//*[@id="container"]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[1]')
        ram.append(ra.text.split('|')[0])
    except NoSuchElementException:
        ram.append("-")
     
    try:
        ro=driver.find_element_by_xpath('//*[@id="container"]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[1]')
        rom.append(ro.text.split('|')[1])
    except NoSuchElementException:
        rom.append("-")
    
    try:
        prim=driver.find_element_by_xpath('//*[@id="container"]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[3]')
        primary.append(prim.text.split('|')[0])
    except NoSuchElementException:
        primary.append('-')
    
    try:
        secon=driver.find_element_by_xpath('//*[@id="container"]/div/div[3]/div[1]/div[2]/div[9]/div[4]/div/div[2]/div[1]/div[5]/table/tbody/tr[5]/td[2]/ul/li')
        secondary.append(secon.text)
    except NoSuchElementException:
        secondary.append('-')
    
    try:
        disp=driver.find_element_by_xpath('//*[@id="container"]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[2]')
        display.append(disp.text)
    except NoSuchElementException:
        display.append('-')
    
    try:
        batt=driver.find_element_by_xpath('//*[@id="container"]/div/div[3]/div[1]/div[2]/div[8]/div[1]/div/div[2]/ul/li[4]')
        battery.append(batt.text)
    except NoSuchElementException:
        battery.append('-')
        
print(len(brand))                     
print(len(price))
print(len(product))
print(len(ram))
print(len(rom))
print(len(colour))
print(len(primary))
print(len(secondary))
print(len(display))
print(len(battery))
```

```python
url = 'https://www.hostelworld.com/'  
driver.get(url)
time.sleep(2)
london=driver.find_element_by_xpath('/html/body/div[3]/div/div/div[1]/div[7]/section/div/div[1]/a')
london.click()
time.sleep(5)
search = driver.find_element_by_xpath('/html/body/div[1]/div[1]/div[2]/section[1]/div/div/form/div/div[2]/button') 
search.click()
```

```python
name = []                        
distance=[]
ratings=[]
reviews=[]
totreview=[]
private=[]
dorms=[]
facilities=[]
hurl=[]

while(True):   
    nam=driver.find_elements_by_xpath('//h2[@class="title title-6"]')                
    dist=driver.find_elements_by_xpath('//div[@class="title-row"]/div/a/span[1]')
    ratn=driver.find_elements_by_xpath('//div[@class="rating rating-summary-container big"]')
    totrev=driver.find_elements_by_xpath('//div[@class="rating rating-summary-container big"]')
    tre=driver.find_elements_by_xpath('//div[@class="rating rating-summary-container big"]')
    priv=driver.find_elements_by_xpath('//div[@class="price-col"]')
    dor=driver.find_elements_by_xpath('//div[@class="price-col"]')
    facility=driver.find_elements_by_xpath('//div[@class="facilities-label facilities"]')
    urls=driver.find_elements_by_xpath("//h2[@class='title title-6']/a")

    for i in nam:
        name.append(i.text)                   
    for j in dist:
        distance.append(j.text)
    for k in ratn:
        ratings.append(k.text.split('\n')[0])
    for l in rev:
        reviews.append(l.text.split('\n')[1])
    for m in totrev:
            treview.append(m.text.split()[2])
    for o in priv:
        private.append(o.text.replace('\n',''))
    for p in dor:
        dorms.append(p.text.replace('\n',''))
    for q in facility:
        facilities.append(q.text)
    for url in urls:
        hurl.append(url.get_attribute('href'))
        
    for n in reviews:                            
        if "Rating"in n:
            reviews[reviews.index(n)] = "None"
        elif "Reviews"in n:
            reviews[reviews.index(n)] = "None"
       
    try:
        next_button = driver.find_element_by_xpath('//div[@class="pagination-item pagination-next"]')  

        next_button.click()

    except:

        break
print(len(name))                                
print(len(distance))
print(len(ratings))
print(len(reviews))
print(len(totreview))
print(len(private))
print(len(dorms))
print(len(facilities))
print(len(hurl))
```
